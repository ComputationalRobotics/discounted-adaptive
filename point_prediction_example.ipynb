```python
# Time Series Functions

import numpy as np
import matplotlib.pyplot as plt

def square_wave(t, T):
    """
    Square wave function
    t: time
    T: period (s)
    """
    return 0.5 * (np.sign(np.sin(2 * np.pi * t / T)) + 1)

# Plot square wave

t = np.linspace(0, 1, 500, endpoint=False)
plt.plot(t, square_wave(t, 0.1))
```

```python
from point_prediction import magnitude_learner

# Create a magnitude learner
learner = magnitude_learner.MagnitudeLearner()

# Create a time series
t = np.linspace(0, 1, 500, endpoint=False)
y = square_wave(t, 0.1)
x = np.zeros_like(y)

for i in range(0,len(t)):
    # Predict
    
    # Update
    if i > 0:
        learner.update(ground_truth=y[i], forecast=x[i-1])
    else:
        learner.update(ground_truth=y[i],forecast=0)
    x[i] = learner.predict()
```

```python
# Plot the results
plt.plot(t, x)
plt.plot(t,y)
```
