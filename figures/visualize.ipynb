```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
# Set the default font to Computer Modern LaTeX
mpl.rcParams['font.family'] = 'cmr10'
mpl.rcParams['font.size'] = 14

# Read the CSV file
df = pd.read_csv('/Users/davidbombara/Documents/ComputationalRoboticsCode/discounted-adaptive/figures/grouped_data_sudden_method.csv')

# Group the data by Method
grouped = df.groupby('Method')

# Create a 2 x 2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot Cov vs D
axs[0, 0].set_title('Cov vs D')
for name, group in grouped:
    axs[0, 0].semilogx(group['D'], group['Cov'], label=name)
axs[0, 0].set_xlabel('D')
axs[0, 0].set_ylabel('Cov')
#axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot Avg Width vs D
axs[0, 1].set_title('Avg Width vs D')
for name, group in grouped:
    axs[0, 1].semilogx(group['D'], group['Avg Width'], label=name)
axs[0, 1].set_xlabel('D')
axs[0, 1].set_ylabel('Avg Width')
#axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot Avg Miscov vs D
axs[1, 0].set_title('Avg Miscov vs D')
for name, group in grouped:
    axs[1, 0].semilogx(group['D'], group['Avg Miscov'], label=name)
axs[1, 0].set_xlabel('D')
axs[1, 0].set_ylabel('Avg Miscov')
#axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot Cov vs D
axs[0, 0].set_title('Cov vs D')
for name, group in grouped:
    axs[0, 0].semilogx(group['D'], group['Cov'], label=name)
axs[0, 0].set_xlabel('D')
axs[0, 0].set_ylabel('Cov')
#axs[0, 0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2)
axs[0, 0].grid(True)

# Plot Avg Width vs D
axs[0, 1].set_title('Avg Width vs D')
for name, group in grouped:
    axs[0, 1].semilogx(group['D'], group['Avg Width'], label=name)
axs[0, 1].set_xlabel('D')
axs[0, 1].set_ylabel('Avg Width')
#axs[0, 1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2)
axs[0, 1].grid(True)

# Plot Avg Miscov vs D
axs[1, 0].set_title('Avg Miscov vs D')
for name, group in grouped:
    axs[1, 0].semilogx(group['D'], group['Avg Miscov'], label=name)
axs[1, 0].set_xlabel('D')
axs[1, 0].set_ylabel('Avg Miscov')
#axs[1, 0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2)
axs[1, 0].grid(True)

# Plot Avg Regret vs D
axs[1, 1].set_title('Avg Regret vs D')
for name, group in grouped:
    axs[1, 1].semilogx(group['D'], group['Avg Regret'], label=name)
    #axs[1, 1].set_ylim(0, 0.025)
axs[1, 1].set_xlabel('D')
axs[1, 1].set_ylabel('Avg Regret')
axs[1, 1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2)
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```
